```python
file_location = "/FileStore/tables/Ecommerce_Customers-4a7d1"
df = spark.read.format("csv").option("inferSchema",True).option("header", True).load(file_location)
display(df)
```

```python
df = df.withColumnRenamed("Avg. Session Length", "avg_session_length")
df = df.withColumnRenamed("Time on App", "time_on_app")
df = df.withColumnRenamed("Time on Website", "time_on_website")
df = df.withColumnRenamed("Length of Membership", "length_of_membership")
df = df.withColumnRenamed("Yearly Amount Spent", "yearly_amount_spent")
display(df)
```

```python
df.write.save('/FileStore/parquet/assignment_ecommerce',  
               format='parquet') #includes metadata
df = spark.read.load("/FileStore/parquet/assignment_ecommerce")
display(df)
```

```python
df.createOrReplaceTempView("ecom_stats")#create a  view
```

```python
display(spark.sql("""select Email, Address, time_on_app, yearly_amount_spent from ecom_stats order by 4 desc limit 30"""))
```

```python
df_analyze = spark.sql("""select avg_session_length, time_on_website, time_on_app, length_of_membership, yearly_amount_spent from ecom_stats order by 5 desc limit 100""")
display(df_analyze)
```

```python
from pyspark.ml.feature import VectorAssembler, MinMaxScaler
from pyspark.ml.regression import LinearRegression
```

```python
assembler = VectorAssembler(inputCols=['avg_session_length', 'time_on_website', 'time_on_app', 'length_of_membership'],outputCol="features")
transformed = assembler.transform(df_analyze)
scaler = MinMaxScaler(inputCol="features",\
         outputCol="scaledFeatures")
scalerModel =  scaler.fit(transformed.select("features"))
scaledData = scalerModel.transform(transformed)
```

```python
lr = LinearRegression(featuresCol = 'features', labelCol='yearly_amount_spent')
lr_model = lr.fit(scaledData)
```

```python
trainingSummary = lr_model.summary
print("Coefficients: " + str(lr_model.coefficients))
print("RMSE: %f" % trainingSummary.rootMeanSquaredError)
print("R2: %f" % trainingSummary.r2)
```

```python
import matplotlib.pyplot as plt
ax = sns.pairplot(df_visualize)
display(ax.fig)
```
